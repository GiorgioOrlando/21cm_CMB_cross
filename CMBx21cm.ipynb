```python
# low-memory definition of Wigner symbol 3j, valid when m1 = m2 = m3 = 0

def wigner3jm0(l1,l2,l3):
    if l3 >l1+l2 or l3 < np.abs(l1-l2) or (l1+l2+l3) % 2== 1:
        value = 0.
        return value
    else:
        factor1 = mpmath.factorial(l1/2+l2/2+l3/2)
        factor2 = mpmath.factorial(-l1/2+l2/2+l3/2)
        factor3 = mpmath.factorial(l1/2-l2/2+l3/2)
        factor4 = mpmath.factorial(l1/2+l2/2-l3/2)
    
        sq1 = mpmath.sqrt(mpmath.factorial(-l1+l2+l3))
        sq2 = mpmath.sqrt(mpmath.factorial(l1-l2+l3))
        sq3 = mpmath.sqrt(mpmath.factorial(l1+l2-l3))
    
        sq4 = mpmath.sqrt(mpmath.factorial(l1+l2+l3+1))
    
        value = (-1)**(-l1/2-l2/2-l3/2)*factor1*sq1*sq2*sq3/factor2/factor3/factor4/sq4
    
        return float(value)

# def wigner3jm0(l1,l2,l3):
#     return w3j(l1,l2,l3,0,0,0)
```

```python
#functions calling CMB PS with correct normalization, in the end we take r = 0.01, nothing really changes

def cTT(i,r):
    l=llist[i]
    if r == 0.01:
        val = tcmb2[i]/(10**12*l*(l+1)/(2*np.pi)*2.7255**2) 
    elif r == 0.001:
        val = tcmb3[i]/(10**12*l*(l+1)/(2*np.pi)*2.7255**2)
    return val


def cEE(i,r):
    l=llist[i]
    if r == 0.01:
        val = ecmb2[i]/(10**12*l*(l+1)/(2*np.pi)*2.7255**2) 
    elif r == 0.001:
        val = ecmb3[i]/(10**12*l*(l+1)/(2*np.pi)*2.7255**2)
    return val


def cET(i,r):
    l=llist[i]
    if r == 0.01:
        val = tecmb2[i]/(10**12*l*(l+1)/(2*np.pi)*2.7255**2) 
    elif r == 0.001:
        val = tecmb3[i]/(10**12*l*(l+1)/(2*np.pi)*2.7255**2)
    return val

# collecting CMB spectra
cTTspectra = np.array([cTT(i,0.01) for i in range(len(llist))])
cEEspectra = np.array([cEE(i,0.01) for i in range(len(llist))])
cETspectra = np.array([cET(i,0.01) for i in range(len(llist))])
```

# DEFINE USEFUL FUNCTIONS 

```python
# Hubble parameter H(z)
def H(z):
    value = results.h_of_z(z)
    return value

# conformal time \eta(z)
def eta(z):
    value = results.conformal_time(z)
    return value

# co-moving distance r(z)
def r(z):
    integrand =  lambda x: 1/H(x)
    int = scint.quad(integrand,0,z)
    return int[0]

# compute co-moving distance at selected red-shifts and recomb
r30 = r(30)
r50 = r(50)
r100 = r(100)
r1100 = r(1100)

# scale factor a(z)
def a(z):
    value = 1/(1+z)
    return value

# dimensionless scalar PS
def scalarpower(q):
    value = 2*np.pi**2*scal_pow(q)/q**3
    return value

# 21-cm transfer from Munoz et. al. arXiv:1506.04152
def trans(qpar,l,r,z):
    if z==30:
        alpha = alp1r(r30)
        T = t21r(r30)
    elif z==50:
        alpha = alp1r(r50)
        T = t21r(r50)
    elif z==100:
        alpha = alp1r(r100)
        T = t21r(r100)
    value = (alpha + T*qpar**2/(qpar**2+l**2/r**2))
    return value

# Omega_m
Omegam = 0.0223828+0.1201075

# Window function 21-cm Fourier space
def Wq(qpar,z,dnu):
    deltaD = 17*np.sqrt((1+z)/10*0.15/Omegam)*dnu/2
    value = np.exp(-qpar**2*deltaD**2/2)
    return value


# radial direction resolution
def deltar(z,dnu):
    return np.sqrt((1+z)/10*0.15/Omegam)*(dnu/0.1)

# Window function 21-cm
def W(rmean,r,z,dnu):
    deltar = 17*np.sqrt((1+z)/10*0.15/Omegam)*dnu
    value = 1/deltar/np.sqrt(2*np.pi)*np.exp(-1/2*(r-rmean)**2/deltar**2)
    return value
```

## PS, flat-sky exact vs full-sky Limber

```python
# list of l for plot

llistpl = np.logspace(1,5,100)

def trans(qpar,l,r,z):
    if z==30:
        alpha = alp1r(r30)
        T = t21r(r30)
    elif z==50:
        alpha = alp1r(r50)
        T = t21r(r50)
    elif z==100:
        alpha = alp1r(r100)
        T = t21r(r100)
    value = (alpha + T*qpar**2/(qpar**2+l**2/r**2))
    return value

def trans_novel(qpar,l,r,z):
    if z==30:
        alpha = alp1r(r30)
        T = t21r(r30)
    elif z==50:
        alpha = alp1r(r50)
        T = t21r(r50)
    elif z==100:
        alpha = alp1r(r100)
        T = t21r(r100)
    value = (alpha)
    return value
```

```python
# 21-cm PS exact flat sky

def c21(i,r,z,dnu):
    l= int(llistpl[i])
    
    if z == 30:
        integrand = lambda q: Wq(q,z,dnu)**2*trans(q,l,r,z)**2*(mono30f(np.sqrt(q**2+l**2/r**2)))**2*scalarpower(np.sqrt(q**2+l**2/r**2))
    elif z == 50:
        integrand = lambda q: Wq(q,z,dnu)**2*trans(q,l,r,z)**2*(mono50f(np.sqrt(q**2+l**2/r**2)))**2*scalarpower(np.sqrt(q**2+l**2/r**2))
    elif z == 100:
        integrand = lambda q: Wq(q,z,dnu)**2*trans(q,l,r,z)**2*(mono100f(np.sqrt(q**2+l**2/r**2)))**2*scalarpower(np.sqrt(q**2+l**2/r**2))
    
    inte = scint.quad(integrand,1e-5,200.,epsabs=0.,epsrel=1e-4, limit = 10000000)[0]
    factor = l**2/(2*np.pi)
    value = inte*factor/r**2/(2*np.pi)
    return value

def c21_novel(i,r,z,dnu):
    l= int(llistpl[i])
    
    if z == 30:
        integrand = lambda q: Wq(q,z,dnu)**2*trans_novel(q,l,r,z)**2*(mono30f(np.sqrt(q**2+l**2/r**2)))**2*scalarpower(np.sqrt(q**2+l**2/r**2))
    elif z == 50:
        integrand = lambda q: Wq(q,z,dnu)**2*trans_novel(q,l,r,z)**2*(mono50f(np.sqrt(q**2+l**2/r**2)))**2*scalarpower(np.sqrt(q**2+l**2/r**2))
    elif z == 100:
        integrand = lambda q: Wq(q,z,dnu)**2*trans_novel(q,l,r,z)**2*(mono100f(np.sqrt(q**2+l**2/r**2)))**2*scalarpower(np.sqrt(q**2+l**2/r**2))
    
    inte = scint.quad(integrand,1e-5,200.,epsabs=0.,epsrel=1e-4, limit = 10000000)[0]
    factor = l**2/(2*np.pi)
    value = inte*factor/r**2/(2*np.pi)
    return value

c21(0,r100,100.,1.), c21_novel(0,r100,100.,1.)
```

```python
# y-sampling for 21-cm PS computation- full sky Limber 

def Limber21_c21_30(l):
    value = y30*W(r30,y30,30,0.1)**2*mono30f(l/y30)**2*alp1r(y30)**2*scal_pow(l/y30)
    return value   

def Limber21_c21_50(l):
    value = y50*W(r50,y50,50,0.1)**2*mono50f(l/y50)**2*alp1r(y50)**2*scal_pow(l/y50)
    return value   

def Limber21_c21_100(l):
    value = y100*W(r100,y100,100,0.1)**2*mono100f(l/y100)**2*alp1r(y100)**2*scal_pow(l/y100)
    return value   

# y-sampling for 21-cm PS computation- full sky Limber - Deltaw = 1 MHz case  

def Limber21_c21_30_1Mhz(l):
    value = y301*W(r30,y301,30,1)**2*mono30f(l/y301)**2*alp1r(y301)**2*scal_pow(l/y301)
    return value   

def Limber21_c21_50_1Mhz(l):
    value = y501*W(r50,y501,50,1)**2*mono50f(l/y501)**2*alp1r(y501)**2*scal_pow(l/y501)
    return value   

def Limber21_c21_100_1Mhz(l):
    value = y1001*W(r100,y1001,100,1)**2*mono100f(l/y1001)**2*alp1r(y1001)**2*scal_pow(l/y1001)
    return value 
```

```python
# 21-cm PS, limber+no velocity terms, for Fisher forecasts

def c21_30(i,r,z,dnu):
    l= int(llist2[i])
    cms = Limber21_c21_30(l)
    I2= integrate.simps(cms,x=y30)
    factor = 2*np.pi**2/l**3
    value = I2*factor
    return value

def c21_50(i,r,z,dnu):
    l= int(llist2[i])
    cms = Limber21_c21_50(l)
    I2= integrate.simps(cms,x=y50)
    factor = 2*np.pi**2/l**3
    value = I2*factor
    return value

def c21_100(i,r,z,dnu):
    l= int(llist2[i])
    cms = Limber21_c21_100(l)
    I2= integrate.simps(cms,x=y100)
    factor = 2*np.pi**2/l**3
    value = I2*factor
    return value
```

```python
def Fisher_T_30(l1,l2,l3):
    D = 1
    if l1 == l2:
        D=2
    B1 = B_prim_I1_T_30(l1,l2,l3)
    B2 = B_sec_I1_T_30(l1,l2,l3)
    C = C21r30w01[l1-10000]*C21r30w01[l2-10000]*cTTspectra[l3-2]*D
    fish1 = B1**2 / C
    fish2 = B2**2 / C
    fish12= B1*B2 / C
    return fish1, fish2, fish12

def Fisher_T_50(l1,l2,l3):
    D = 1
    if l1 == l2:
        D=2
    B1 = B_prim_I1_T_50(l1,l2,l3)
    B2 = B_sec_I1_T_50(l1,l2,l3)
    C = C21r50w01[l1-10000]*C21r50w01[l2-10000]*cTTspectra[l3-2]*D
    fish1 = B1**2 / C
    fish2 = B2**2 / C
    fish12= B1*B2 / C
    return fish1, fish2, fish12

def Fisher_T_100(l1,l2,l3):
    D = 1
    if l1 == l2:
        D=2
    B1 = B_prim_I1_T_100(l1,l2,l3)
    B2 = B_sec_I1_T_100(l1,l2,l3)
    C = C21r100w01[l1-10000]*C21r100w01[l2-10000]*cTTspectra[l3-2]*D
    fish1 = B1**2 / C
    fish2 = B2**2 / C
    fish12= B1*B2 / C
    return fish1, fish2, fish12

from tqdm import tqdm

def Fisher_l3(Fisher,l3,llist2):
    # wig.wig_table_free()
    # wig.wig_temp_free()
    # wig.wig_table_init(200002,6)
    # wig.wig_temp_init(200002)
    val = np.zeros((len(llist2),2,2))
    for i1 in tqdm(range(len(llist2)),disable=(l3!=l3list[-1])):
    # for i1 in tqdm(range(len(llist2))):
        l1 = llist2[i1]
        # l2list = np.arange(l1,np.fmin(l1+l3,llist2[-1])+1)
        l2list = np.arange(np.fmax(l1-l3,llist2[0]),l1+1)
        for l2 in l2list:
            if (l2+l1+l3) % 2== 0:
                fishy = Fisher(l2,l1,l3) 
                val[i1,0,0] += fishy[0]
                val[i1,1,1] += fishy[1]
                val[i1,0,1] += fishy[2]
    # wig.wig_table_free()
    # wig.wig_temp_free()
    val[:,1,0] = val[:,0,1]
    return val.cumsum(0)
```

```python
# Fisher terms 21-21-T + 21-21-E primordial

def Fisher_TE_30(l1,l2,l3): 
    D = 1
    if l1 == l2:
        D=2
        
    B1_T = B_prim_I1_T_30(l1,l2,l3)
    B1_E = B_prim_I1_E_30(l1,l2,l3)

    B2_T = B_sec_I1_T_30(l1,l2,l3)
    B2_E = B_sec_I1_E_30(l1,l2,l3)

    cterm = (cTTspectra[l3-2]*cEEspectra[l3-2]-cETspectra[l3-2]**2)
    CTT = C21r30w01[l1-10000]*C21r30w01[l2-10000]*cterm*D / cEEspectra[l3-2]
    CEE = C21r30w01[l1-10000]*C21r30w01[l2-10000]*cterm*D / cTTspectra[l3-2]
    CET = C21r30w01[l1-10000]*C21r30w01[l2-10000]*cterm*D / cETspectra[l3-2]
    
    fish1 = B1_T**2 / CTT
    fish1+= B1_E**2 / CEE
    fish1+= -2*B1_T*B1_E / CET

    fish2 = B2_T**2 / CTT
    fish2+= B2_E**2 / CEE
    fish2+= -2*B2_T*B2_E / CET

    fish12 = B1_T*B2_T / CTT
    fish12+= B1_E*B2_E / CEE
    fish12+= -B1_T*B2_E / CET
    fish12+= -B1_E*B2_T / CET
    
    return fish1, fish2, fish12

# Fisher terms 21-21-T + 21-21-E primordial

def Fisher_TE_50(l1,l2,l3): 
    D = 1
    if l1 == l2:
        D=2
        
    B1_T = B_prim_I1_T_50(l1,l2,l3)
    B1_E = B_prim_I1_E_50(l1,l2,l3)

    B2_T = B_sec_I1_T_50(l1,l2,l3)
    B2_E = B_sec_I1_E_50(l1,l2,l3)

    cterm = (cTTspectra[l3-2]*cEEspectra[l3-2]-cETspectra[l3-2]**2)
    CTT = C21r50w01[l1-10000]*C21r50w01[l2-10000]*cterm*D / cEEspectra[l3-2]
    CEE = C21r50w01[l1-10000]*C21r50w01[l2-10000]*cterm*D / cTTspectra[l3-2]
    CET = C21r50w01[l1-10000]*C21r50w01[l2-10000]*cterm*D / cETspectra[l3-2]
    
    fish1 = B1_T**2 / CTT
    fish1+= B1_E**2 / CEE
    fish1+= -2*B1_T*B1_E / CET

    fish2 = B2_T**2 / CTT
    fish2+= B2_E**2 / CEE
    fish2+= -2*B2_T*B2_E / CET

    fish12 = B1_T*B2_T / CTT
    fish12+= B1_E*B2_E / CEE
    fish12+= -B1_T*B2_E / CET
    fish12+= -B1_E*B2_T / CET
    
    return fish1, fish2, fish12

# Fisher terms 21-21-T + 21-21-E primordial

def Fisher_TE_100(l1,l2,l3): 
    D = 1
    if l1 == l2:
        D=2
        
    B1_T = B_prim_I1_T_100(l1,l2,l3)
    B1_E = B_prim_I1_E_100(l1,l2,l3)

    B2_T = B_sec_I1_T_100(l1,l2,l3)
    B2_E = B_sec_I1_E_100(l1,l2,l3)

    cterm = (cTTspectra[l3-2]*cEEspectra[l3-2]-cETspectra[l3-2]**2)
    CTT = C21r100w01[l1-10000]*C21r100w01[l2-10000]*cterm*D / cEEspectra[l3-2]
    CEE = C21r100w01[l1-10000]*C21r100w01[l2-10000]*cterm*D / cTTspectra[l3-2]
    CET = C21r100w01[l1-10000]*C21r100w01[l2-10000]*cterm*D / cETspectra[l3-2]
    
    fish1 = B1_T**2 / CTT
    fish1+= B1_E**2 / CEE
    fish1+= -2*B1_T*B1_E / CET

    fish2 = B2_T**2 / CTT
    fish2+= B2_E**2 / CEE
    fish2+= -2*B2_T*B2_E / CET

    fish12 = B1_T*B2_T / CTT
    fish12+= B1_E*B2_E / CEE
    fish12+= -B1_T*B2_E / CET
    fish12+= -B1_E*B2_T / CET
    
    return fish1, fish2, fish12
```
